```python
import numpy as np
import scipy
import matplotlib.pyplot as plt 
import math

plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

%matplotlib inline
# for auto-reloading external modules
# see http://stackoverflow.com/questions/1907993/autoreload-of-modules-in-ipython
%load_ext autoreload
%autoreload 2
```

```python
def actFunction(I):
    if I <= 0:
        I = 0
    return I

def sigmFunction(I):
    y = (1 + np.tanh(I)) / 2
    return y

def actVect(vec, func):
    func = np.vectorize(func)
    return func(vec)

def dIdt(W, I, E, y, tau):
    return (np.dot(W, y) - I + E) / tau
 
def attractorNetwork(tt, dt, w, v, E, tau, act=actFunction):
    W = np.matrix([[w, v], [v, w]])
    Y = np.zeros([2, len(tt)])
    I = np.zeros([2, len(tt)])
    I[:, 0] = [2, 0]
    Y[:, 0] = actVect(I[:, 0], actFunction)
    for i in range(1, len(tt)):
        I[:, i] = I[:, i-1] + dIdt(W, I[:, i-1], E[i-1], Y[:, i-1], tau) * dt
        Y[:, i] = actVect(I[:, i], act)
    Y1 = Y[0, :]
    Y2 = Y[1, :]
    return Y1, Y2, I
```

```python
#simulate the system for w = 0.2, v = 0.8 with no external input
dt = 0.01
tEnd = 10
tt = np.arange(0, 10, 0.01)
tau = 0.2
w = 0.2
v = 0.8
E = np.zeros_like(tt)
Y1, Y2, I = attractorNetwork(tt, dt, w, v, E, tau)

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 1.2, 1.2])
axes.grid(True)
axes.plot(tt, Y1, label='y1')
axes.plot(tt, Y2, label='y2')
axes.set_xlabel('Time')
axes.set_title('Attractor Network (w = 0.2, v = 0.8)')
axes.legend()
```
