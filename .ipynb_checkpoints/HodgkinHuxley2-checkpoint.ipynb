```python
dt = 0.001
t1 = 0
t2 = 10 #time when ramping stops
tt = np.arange(t1, t2, dt)

rampValue = 10 #max value achieved after ramping
rampStep = rampValue / len(tt)
I = np.zeros_like(tt)

for t in range(len(tt) - 1):
    I[t+1] = I[t] + rampStep 

h, m, n, u = hodgkin_huxley(tt, I)
        
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 1, 1])
axes.plot(tt, n, 'b', label='n')
axes.plot(tt, h, 'r', label='h')
axes.plot(tt, m, 'k', label='m')
axes.set_xlabel('Time [ms]')
axes.set_title('Gating variables')
axes.legend()

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 1, 1])
axes.plot(tt, u, label='Transmembrane voltage (u)')
axes.set_xlabel('Time [ms]')
axes.set_ylabel('Voltage [mV]')
axes.legend()

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 1, 1])
axes.plot(tt, I, label='Input Current (I)')
axes.set_xlabel('Time [ms]')
axes.set_ylabel('Current')
axes.legend()
```

```python
dt = 0.001
t1 = 0
tRamp = 25 #time when ramping stops
t2 = 50.
tt = np.arange(t1, t2, dt)
ttRamp = np.arange(t1, tRamp, dt)
endRamp = tRamp / dt

rampValue = 10 #max value achieved after ramping
rampStep = rampValue / len(ttRamp)
I = np.ones_like(tt) * rampValue
I[0:len(ttRamp)] = 0

for t in range(len(ttRamp) - 1):
    I[t+1] = I[t] + rampStep 

h, m, n, u = hodgkin_huxley(tt, I)
        
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 1, 1])
axes.plot(tt, n, 'b', label='n')
axes.plot(tt, h, 'r', label='h')
axes.plot(tt, m, 'k', label='m')
axes.set_xlabel('Time [ms]')
axes.set_title('Gating variables')
axes.legend()

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 1, 1])
axes.plot(tt, u, label='Transmembrane voltage (u)')
axes.set_xlabel('Time [ms]')
axes.set_ylabel('Voltage [mV]')
axes.legend()

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 1, 1])
axes.plot(tt, I, label='Input Current (I)')
axes.set_xlabel('Time [ms]')
axes.set_ylabel('Current')
axes.legend()
```

```python
t1 = 0
t2 = 500.
dt = 0.001
tt = np.arange(t1, t2, dt)
I = 10 * np.ones_like(tt)

h, m, n, u = hodgkin_huxley(tt, I)
peaks = detect_peaks(u) 
firingRate = (float(len(peaks))/t2) * 1000 #number of spikes per SECOND
print(firingRate)

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 1, 1])
axes.plot(tt, n, 'b', label='n')
axes.plot(tt, h, 'r', label='h')
axes.plot(tt, m, 'k', label='m')
axes.set_xlabel('Time [ms]')
axes.set_title('Gating variables')
axes.legend()

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 1, 1])
axes.plot(tt, u, label='Transmembrane voltage (u)')
axes.plot(tt[peaks], u[peaks], marker='o', ls='')
axes.set_xlabel('Time [ms]')
axes.set_ylabel('Voltage [mV]')
axes.legend()
```

```python
t1 = 0
t2 = 50.
dt = 0.001
tt = np.arange(t1, t2, dt)
currents = np.linspace(0, 20, 100)
firingRate = np.zeros_like(currents)
for i in range(len(currents)):
    I = currents[i] * np.ones_like(tt)
    h, m, n, u = hodgkin_huxley(tt, I)
    peaks = detect_peaks(u, mpd=0) 
    firingRate[i] = (float(len(peaks))/t2) * 1000 #number of spikes per SECOND
    
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 1, 1])
axes.plot(currents, firingRate, label='Activation Function')
axes.set_xlabel('Input Current')
axes.set_ylabel('Firing Rate')
axes.legend()
```

```python
dt = 0.001
t1 = 0
t2 = 2000
tt = np.arange(t1, t2, dt)

I = 15 + np.sin(2*math.pi*0.001*tt)

h, m, n, u = hodgkin_huxley(tt, I)

```

```python
peaks = detect_peaks(u)
firingRate = np.zeros(len(peaks) - 1)
for i in range(len(peaks) - 1):
    firingRate[i] = 1. / (tt[peaks[i + 1]] - tt[peaks[i]])
firingRate *= 1000

    
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 2, 2])
axes.plot(tt, n, 'b', label='n')
axes.plot(tt, h, 'r', label='h')
axes.plot(tt, m, 'k', label='m')
axes.set_xlabel('Time [ms]')
axes.set_title('Gating variables')
axes.legend()

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 2, 2])
axes.plot(tt, u, label='Transmembrane voltage (u)')
axes.plot(tt[peaks], u[peaks], marker='o', ls='')
axes.set_xlabel('Time [ms]')
axes.set_ylabel('Voltage [mV]')
axes.legend()

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 1, 1])
axes.plot(tt, I, label='Input Current (I)')
axes.set_xlabel('Time [ms]')
axes.set_ylabel('Current')
axes.legend()

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 1, 1])
axes.plot(tt[peaks[1:]], firingRate)
axes.set_xlabel('Time [ms]')
axes.set_ylabel('Firing Rate [Hz]')
```

```python
dt = 0.001
t1 = 0
t2 = 8000
tt = np.arange(t1, t2, dt)

I = 5 + 5 * np.sin(2*math.pi*0.001*tt)

h, m, n, u = hodgkin_huxley(tt, I)
h_n, m_n, n_n, u_n = hodgkin_huxley_noise(tt, I)
```

```python
peaks = detect_peaks(u, mph=20)
peaks_n = detect_peaks(u_n, mph=20)
firingRate = np.zeros(len(peaks) - 1)
for i in range(len(peaks) - 1):
    firingRate[i] = 1. / (tt[peaks[i+1]] - tt[peaks[i]])
firingRate *= 1000

firingRateN = np.zeros(len(peaks_n) - 1)
for i in range(len(peaks_n) - 1):
    firingRateN[i] = 1. / (tt[peaks_n[i+1]] - tt[peaks_n[i]])
firingRateN *= 1000
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 1, 1])
axes.plot(tt, I, label='Input Current (I)')
axes.set_xlabel('Time [ms]')
axes.set_ylabel('Current')
axes.legend()

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 2, 2])
axes.plot(tt, u, label='Transmembrane voltage (u)')
axes.plot(tt[peaks], u[peaks], marker='o', ls='')
axes.set_xlabel('Time [ms]')
axes.set_ylabel('Voltage [mV]')
axes.legend()

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 2, 2])
axes.plot(tt, u_n, label='Transmembrane voltage [NOISE] (u)')
axes.plot(tt[peaks_n], u[peaks_n], marker='o', ls='')
axes.set_xlabel('Time [ms]')
axes.set_ylabel('Voltage [mV]')
axes.legend()
```

The model with noise is actually functional and encodes the signal variation. Whereas the model without channel noise doesn't encode the signal in its firing rate. Therefore channel noise is actually a property that is important for the neuron to work in a proper way.

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 1, 1])
axes.plot(tt[peaks[1:]], firingRate)
axes.set_xlabel('Time [ms]')
axes.set_ylabel('Firing Rate [Hz]')

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 1, 1])
axes.plot(tt[peaks_n[1:]], firingRateN)
axes.set_xlabel('Time [ms]')
axes.set_ylabel('Firing Rate (NOISE) [Hz]')
```
